```python
import pandas
df = pandas.read_csv("supermarkets.csv")
```

```python
df
```

```python
df["Address"] = df["Address"]+","+df["City"]+","+df["Country"] #Add city and country column to Address column.
```

```python
df
```

```python
df["Coordinates"] = df["Address"].apply(nom.geocode) #Create a new column called coordinates for geocode output.
df
```

```python
df.Coordinates[0].latitude
```

```python
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x!=None else None) #New column for latitude.
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude if x!=None else None) #New column for longitude.
df
```
